```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
titanic.tail()
titanic.info()
titanic.describe()

# Each column in a DataFrame is a Series
type(titanic["Age"])
titanic[["Age", "Sex"]].shape 
titanic[["Age", "Sex"]].info()
```

```python
""" Filtering specific rows from a dataframe """
titanic[titanic["Sex"] == "female"].head()
titanic[titanic["Sex"] == "male"].head()
titanic[titanic["Sex"] == "male"].tail()
titanic[titanic["Sex"] == "female"].tail()
titanic[titanic["Sex"] == "male"].shape
titanic[titanic["Sex"] == "female"].shape

above_35 = titanic[titanic["Age"] > 35]
above_35.head()
above_35.shape
bool_output = titanic["Age"] > 35

# .isin()
class_23_one = titanic[titanic["Pclass"].isin([2, 3])] #or
class_23_two = titanic[(titanic["Pclass"]==2) | (titanic["Pclass"]==3)] 
print(class_23_one.head())
print(class_23_two.head())

# The notna() conditional function returns a True for each row the values are not an Null value.
age_no_na = titanic[titanic['Age'].notna()]
age_no_na.head()
```

```python
""" How do I select specific rows and columns from a DataFrame? 
    -Using loc(label based indexing) and iloc(integer based labelling)
"""
adult_names = titanic.loc[titanic["Age"] > 35, ["Name","Age"]]
adult_names_two = titanic.iloc[9:25, 2:5]

adult_names.head()
adult_names_two.head()

titanic.iloc[0:0, 3] = "anonymous"
print(titanic.iloc[[0, 2], [1, 2]])
titanic.head().head()



# adult_names.shape
```
