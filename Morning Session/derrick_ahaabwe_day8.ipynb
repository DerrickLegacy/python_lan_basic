```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("niceguys.csv")

# Clean the data
data.columns = data.columns.str.strip().str.replace(" ", "_")
data = data.rename(columns={"do_you_have_aboyfriend/girlfriend": "has_bf_gf"})
data["living_allowance"] = data["living_allowance"].str.replace(",", "").astype(float)
```

```python
""" Question 1 Solution """
# Create the histogram, pie, densith chart  using plot for those with bf or bf

# a) Pie Chart
filtered_data_with_bg_gf = data.loc[data["has_bf_gf"] == 1, 'age']
ax = filtered_data_with_bg_gf.value_counts().plot(kind="pie", figsize=(10, 6), autopct='%1.1f%%', startangle=180)
plt.title('Age Distribution of Students with a Boyfriend/Girlfriend')
plt.ylabel('') 
plt.show()

filtered_data_without_bg_gf = data.loc[data["has_bf_gf"] == 0, 'age']
ay = filtered_data_without_bg_gf.value_counts().plot(kind="pie", figsize=(10, 6), autopct='%1.1f%%', startangle=180)
plt.title('Age Distribution of Students without a Boyfriend/Girlfriend')
plt.ylabel('') 
plt.show()
```

```python
# b) Density chart
filtered_data_with_bf_gf = data.loc[data["has_bf_gf"] == 1, 'age']
filtered_data_without_bf_gf = data.loc[data["has_bf_gf"] == 0, 'age']

ax = filtered_data_with_bf_gf.value_counts().plot(kind="density", figsize=(10, 6))
ax = filtered_data_without_bf_gf.value_counts().plot(kind="density", figsize=(10, 6))

ax.axvline(filtered_data_with_bf_gf.mean(), color='red', linestyle='--', linewidth=2, label='Mean Age (Has BF/GF)')
ax.axvline(filtered_data_without_bf_gf.mean(), color='green', linestyle='--', linewidth=2, label='Mean Age (No BF/GF)')

plt.title('Age Distribution of Students with or without a Boyfriend/Girlfriend')
plt.xlabel('Age')
plt.ylabel('Count',loc='bottom')
plt.legend(title="Has BF/GF", labels=["Yes", "No"])
plt.grid(True)
plt.show()
```

```python
# c. Histogram

filtered_data_with_bf_gf = data.loc[data["has_bf_gf"] == 1, 'age']
filtered_data_without_bf_gf = data.loc[data["has_bf_gf"] == 0, 'age']

# Create a combined DataFrame for Seaborn
combined_data = data[data['has_bf_gf'].isin([0, 1])]

plt.figure(figsize=(10, 6))
sb.histplot(data=combined_data, x='age', hue='has_bf_gf', multiple='layer', palette=['orange', 'red'], alpha=0.5, kde=True)

# Add titles and labels
plt.title('Age Distribution Histogram Chart of Students with or without a Boyfriend/Girlfriend')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title="Has BF/GF", labels=["No BF/GF", "Has BF/GF"])
plt.grid(True)
plt.show()

""" 
Notes

-Kernel Density Estimation is a method for estimating the probability density function of a continuous random variable. In simpler terms, it creates a smooth curve that represents the distribution of data points."""
```

```python
# c. Bar Chart
filtered_data_with_bf_gf = data.loc[data["has_bf_gf"] == 1, 'age']
filtered_data_without_bf_gf = data.loc[data["has_bf_gf"] == 0, 'age']

print("With Bf: ",filtered_data_with_bf_gf.count())
print("Without Bf: ",filtered_data_without_bf_gf.count())

ax = filtered_data_with_bf_gf.value_counts().sort_index().plot(kind="bar", figsize=(10, 6), color='red', alpha=0.5, label='Has BF/GF')
filtered_data_without_bf_gf.value_counts().sort_index().plot(kind="bar", figsize=(10, 6), color='orange', alpha=0.5, label='No BF/GF', ax=ax)

plt.title('Age Distributionhart Bar Chart of Students with or without a Boyfriend/Girlfriend')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title="Has BF/GF")
plt.grid(True)
plt.show()
```

```python
# c. Line Chart
filtered_data_with_bf_gf = data.loc[data["has_bf_gf"] == 1, 'age']
filtered_data_without_bf_gf = data.loc[data["has_bf_gf"] == 0, 'age']

print("With Bf: ",filtered_data_with_bf_gf.count())
print("Without Bf: ",filtered_data_without_bf_gf.count())

ax = filtered_data_with_bf_gf.value_counts().sort_index().plot(kind="line", figsize=(10, 6), color='red', alpha=0.5, label='Has BF/GF')
filtered_data_without_bf_gf.value_counts().sort_index().plot(kind="line", figsize=(10, 6), color='orange', alpha=0.5, label='No BF/GF', ax=ax)

plt.title('Age Distributionhart Bar Chart of Students with or without a Boyfriend/Girlfriend')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title="Has BF/GF")
plt.grid(True)
plt.show()
```

```python
""" Question 2 Solution :  Gender who have boyfriend and girlfriend"""

# Filter the data for males and females who have a boyfriend/girlfriend
gender_with_bf_gf = data.loc[(data['sex'].isin(['male', 'female'])) & (data["has_bf_gf"] == 1), 'age']
age_counts = gender_with_bf_gf.value_counts()
print(age_counts)
ax = age_counts.plot(kind='bar', figsize=(6, 4), color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Age Distribution of Students who have Boyfriend/Girlfriend')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True, axis='y')

plt.show()
```
